```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = "/content/data-spell-checker.xlsx"
data = pd.read_excel(file_path)

# Filter correct words where label is 1
correct_words = data[data['label'] == 1]['word'].str.strip().tolist()

# Build vocabulary from correct words
vocab = set(correct_words)
print(f"Vocabulary Size: {len(vocab)}")

# Function to calculate edit distance
def edit_distance(word1, word2):
    dp = np.zeros((len(word1) + 1, len(word2) + 1), dtype=int)
    for i in range(len(word1) + 1):
        dp[i][0] = i
    for j in range(len(word2) + 1):
        dp[0][j] = j

    for i in range(1, len(word1) + 1):
        for j in range(1, len(word2) + 1):
            cost = 0 if word1[i - 1] == word2[j - 1] else 1
            dp[i][j] = min(dp[i - 1][j] + 1,       # Deletion
                           dp[i][j - 1] + 1,       # Insertion
                           dp[i - 1][j - 1] + cost)  # Substitution
    return dp[len(word1)][len(word2)]

# Spell correction function
def correct_word(word, vocab, max_distance=2):
    """
    Corrects a word by finding the closest match from the vocabulary.
    If no match is found within the max_distance, the word is returned as is.
    """
    if word in vocab:  # If the word is already correct
        return word

    closest_word = None
    min_distance = float("inf")
    for v_word in vocab:
        distance = edit_distance(word, v_word)
        if distance < min_distance and distance <= max_distance:
            min_distance = distance
            closest_word = v_word
    return closest_word if closest_word else word

# Function to check and correct words
def spell_check(sentence, vocab, max_distance=2):
    """
    Splits a sentence into words, checks each word against the vocabulary,
    and returns the corrected sentence.
    """
    words = sentence.split()
    corrected_words = [correct_word(word, vocab, max_distance) for word in words]
    return " ".join(corrected_words)

# Example: Test the spell checker
test_sentence = "අංකය් අංකතම අභිචෝදකය්"
corrected_sentence = spell_check(test_sentence, vocab)
print(f"Original: {test_sentence}")
print(f"Corrected: {corrected_sentence}")
```

```python
# Function to calculate accuracy
def calculate_accuracy(true_words, predicted_words):
    """
    Calculates word-level accuracy by comparing true and predicted corrections.
    """
    correct_count = sum([1 for true, pred in zip(true_words, predicted_words) if true == pred])
    return (correct_count / len(true_words)) * 100

# Function to calculate RMSE (average edit distance)
def calculate_average_edit_distance(true_words, predicted_words):
    """
    Calculates the average edit distance between true and predicted words.
    """
    total_distance = sum([edit_distance(true, pred) for true, pred in zip(true_words, predicted_words)])
    return total_distance / len(true_words)

# Example test dataset
test_data = {
    "misspelled": ["අංකය්", "අංකතම", "අභිචෝදකය්"],  # Misspelled words
    "correct": ["අංකය", "අංකතම", "අභිචෝදකයා"]         # Correct spellings
}

# Test the spell checker
predicted_corrections = [correct_word(word, vocab) for word in test_data["misspelled"]]

# Calculate metrics
accuracy = calculate_accuracy(test_data["correct"], predicted_corrections)
average_edit_distance = calculate_average_edit_distance(test_data["correct"], predicted_corrections)

# Print results
print(f"Word-Level Accuracy: {accuracy:.2f}%")
print(f"Average Edit Distance (RMSE): {average_edit_distance:.2f}")
```

```python
import pandas as pd
import numpy as np

# Load the dataset
file_path = "/content/data-spell-checker.xlsx"
data = pd.read_excel(file_path)

# Filter correct words where label is 1
correct_words = data[data['label'] == 1]['word'].str.strip().tolist()

# Build vocabulary from correct words
vocab = set(correct_words)
print(f"Vocabulary Size: {len(vocab)}")

# Function to calculate edit distance
def edit_distance(word1, word2):
    dp = np.zeros((len(word1) + 1, len(word2) + 1), dtype=int)
    for i in range(len(word1) + 1):
        dp[i][0] = i
    for j in range(len(word2) + 1):
        dp[0][j] = j

    for i in range(1, len(word1) + 1):
        for j in range(1, len(word2) + 1):
            cost = 0 if word1[i - 1] == word2[j - 1] else 1
            dp[i][j] = min(dp[i - 1][j] + 1,       # Deletion
                           dp[i][j - 1] + 1,       # Insertion
                           dp[i - 1][j - 1] + cost)  # Substitution
    return dp[len(word1)][len(word2)]

# Spell correction function
def correct_word(word, vocab, max_distance=2):
    """
    Corrects a word by finding the closest match from the vocabulary.
    If no match is found within the max_distance, the word is returned as is.
    """
    if word in vocab:  # If the word is already correct
        return word

    closest_word = None
    min_distance = float("inf")
    for v_word in vocab:
        distance = edit_distance(word, v_word)
        if distance < min_distance and distance <= max_distance:
            min_distance = distance
            closest_word = v_word
    return closest_word if closest_word else word

# Grammar checker function
def grammar_checker(sentence):
    """
    Checks for grammatical consistency based on starting words ('මම', 'අපි').
    """
    words = sentence.split()
    if not words:
        return True, "Empty sentence."

    start_word = words[0]
    end_word = words[-1]
    if start_word == "මම" and not end_word.endswith("මි"):
        return False, "Grammatical error: Sentence starting with 'මම' must end with 'මි'."
    if start_word == "අපි" and not end_word.endswith("මු"):
        return False, "Grammatical error: Sentence starting with 'අපි' must end with 'මු'."
    return True, "No grammatical errors."

# Function to check and correct user input
def spell_check(sentence, vocab, max_distance=2):
    """
    Splits a sentence into words, checks each word against the vocabulary,
    and returns the corrected sentence.
    """
    ignore_words = {"නම", "ගම", "රට"}  # Context-specific words to ignore the next word's spelling
    words = sentence.split()
    corrected_words = []
    skip_next = False

    for i, word in enumerate(words):
        if skip_next:
            corrected_words.append(word)  # Keep the next word as-is
            skip_next = False
            continue

        if word in ignore_words:
            corrected_words.append(word)
            skip_next = True  # Skip correction for the next word
        else:
            corrected_words.append(correct_word(word, vocab, max_distance))

    return " ".join(corrected_words)

# User input loop
print("Sinhala Spell Checker with Grammar Checker")
print("Type 'exit' to quit.")
while True:
    user_input = input("Enter a Sinhala sentence: ").strip()
    if user_input.lower() == "exit":
        print("Goodbye!")
        break

    # Spell check and grammar check
    corrected_output = spell_check(user_input, vocab)
    grammar_ok, grammar_message = grammar_checker(corrected_output)

    # Display results
    print(f"Corrected Output: {corrected_output}")
    if not grammar_ok:
        print(f"Grammar Check: {grammar_message}")
    else:
        print("Grammar Check: Passed")
```
